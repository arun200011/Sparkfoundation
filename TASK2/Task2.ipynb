```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data from the CSV file

data = pd.read_csv("C:\\Users\\arunr\\OneDrive\\Desktop\\globalterrorismdb_0718dist.csv", encoding='latin1')
data.head()
```

```python
# Display the first few rows of the dataset
print("Sample Data:")
print(data.head())
```

```python
# Check the columns in the dataset
print("\nColumns in the Dataset:")
print(data.columns)
```

```python
# Perform data exploration and analysis
print("\nData Exploration:")
print("Total number of rows and columns:", data.shape)
print("Information about the dataset:")
print(data.info())
```

```python
# Filter the dataset for relevant columns
relevant_columns = ['iyear', 'imonth', 'country','country_txt','region_txt', 'city', 'latitude', 'longitude', 'success', 'attacktype1_txt']
filtered_data = data[relevant_columns]
```

```python
# Check for missing values
print("\nMissing Values:")
print(filtered_data.isnull().sum())
```

```python
#droping missing values
filtered_data.dropna()
```

```python
# Analyze hot zones of terrorism
print("\nHot Zones of Terrorism:")
hot_zones = filtered_data['region_txt'].value_counts().head(10)
print(hot_zones)
```

```python
# Visualize hot zones
plt.figure(figsize=(10, 6))
sns.barplot(x=hot_zones.values, y=hot_zones.index, palette='viridis')
plt.xlabel('Number of Terrorist Incidents')
plt.ylabel('Region')
plt.title('Top 10 Hot Zones of Terrorism')
plt.show()
```

```python
# Analyze hot zones of terrorism based on country
print("\nHot Zones of Terrorism:")
hott_zones = filtered_data['country_txt'].value_counts().head(10)
print(hott_zones)
```

```python
# Visualize hot zones of terrorism based on country
plt.figure(figsize=(10, 6))
sns.barplot(x=hott_zones.values, y=hott_zones.index, palette='viridis')
plt.xlabel('Number of Terrorist Incidents')
plt.ylabel('Country')
plt.title('Top 10 Hot Zones of Terrorism')
plt.show()
```
